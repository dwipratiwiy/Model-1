```python
DATASET_PATH='/content/drive/MyDrive/Deep Learning/Exam-Violation-Detection-2'
TRAIN_DIR='/content/drive/MyDrive/Deep Learning/Exam-Violation-Detection-2/train/images'
VALID_DIR='/content/drive/MyDrive/Deep Learning/Exam-Violation-Detection-2/valid/images'
TEST_DIR='/content/drive/MyDrive/Deep Learning/Exam-Violation-Detection-2/test/images'
image_folder = os.path.join(TRAIN_DIR)
image_files = os.listdir(image_folder)

fig, axes = plt.subplots(4, 5, figsize=(15, 10))
for i, img_name in enumerate(image_files[:20]):
    img_path = os.path.join(image_folder, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    row, col = divmod(i, 5)
    axes[row, col].imshow(img)
    axes[row, col].axis("off")
plt.show()
```

```python
train_dir = os.path.join(DATASET_PATH, 'train/images')
valid_dir = os.path.join(DATASET_PATH, 'valid/images')
test_dir = os.path.join(DATASET_PATH, 'test/images')

num_train = len(os.listdir(train_dir)) if os.path.exists(TRAIN_DIR) else 0
num_valid = len(os.listdir(valid_dir)) if os.path.exists(VALID_DIR) else 0
num_test = len(os.listdir(test_dir)) if os.path.exists(TEST_DIR) else 0

num_train = len(os.listdir(train_dir)) if os.path.exists(train_dir) else 0
num_valid = len(os.listdir(valid_dir)) if os.path.exists(valid_dir) else 0
num_test = len(os.listdir(test_dir)) if os.path.exists(test_dir) else 0

categories = ['Train', 'Valid', 'Test']
counts = [num_train, num_valid, num_test]

colors = ['#6b8e23', '#ffcc99', '#4682b4']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(categories, counts, color=colors)
axes[0].set_xlabel("Dataset")
axes[0].set_ylabel("Jumlah Gambar")
axes[0].set_title("Distribusi Jumlah Gambar dalam Dataset")
axes[0].set_ylim(0, max(counts) + 30)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(counts):
    axes[0].text(i, v + 2, str(v), ha='center', fontsize=12, fontweight='bold')

axes[1].pie(counts, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axes[1].set_title("Persentase Distribusi Dataset")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Baca file CSV
df = pd.read_csv("/content/runs/detect/train2/results.csv")

# Konfigurasi style
sns.set(style="whitegrid")

# Buat figure untuk beberapa grafik
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Grafik untuk training loss (gunakan train/box_loss)
sns.lineplot(x=df["epoch"], y=df["train/box_loss"], ax=axes[0, 0], marker="o", label="Train Box Loss")
axes[0, 0].set_title("Training Box Loss per Epoch")

# Grafik untuk validation loss (gunakan val/box_loss)
sns.lineplot(x=df["epoch"], y=df["val/box_loss"], ax=axes[0, 1], marker="o", label="Validation Box Loss")
axes[0, 1].set_title("Validation Box Loss per Epoch")

# Grafik untuk mAP50
sns.lineplot(x=df["epoch"], y=df["metrics/mAP50(B)"], ax=axes[1, 0], marker="o", label="mAP50")
axes[1, 0].set_title("mAP50 per Epoch")

# Grafik untuk Precision dan Recall
sns.lineplot(x=df["epoch"], y=df["metrics/precision(B)"], ax=axes[1, 1], marker="o", label="Precision")
sns.lineplot(x=df["epoch"], y=df["metrics/recall(B)"], ax=axes[1, 1], marker="s", label="Recall")
axes[1, 1].set_title("Precision & Recall per Epoch")
axes[1, 1].legend()

# Tampilkan grafik
plt.tight_layout()
plt.show()
```
